# Price-Range

We sorted the list from the most expensive and also the cheapest, also with the price range from IDR 50000 to 100000. Here is the order from the most expensive.

The most expensive one is TGI Fridays Central Park which priced for at least 120K. Most fast food available are Burger King and Carl's Jr. 

```python
#most expensive
burger_price = sqldf("""SELECT partlor_name, district, unique_menu, price 
                        FROM burger 
                        ORDER BY price DESC""")
burger_price.head().style.hide_index()
```

```python
#cheapest
burger_price = sqldf("""SELECT partlor_name, district, unique_menu, price 
                        FROM burger 
                        ORDER BY price""")
burger_price.head().style.hide_index()
```

```python
#more than IDR 100000
burger_price = sqldf("""SELECT partlor_name, district, unique_menu, price 
                        FROM burger 
                        WHERE price >= 100000""")
burger_price.style.hide_index()
```

```python
#between 50000 to 100000
burger_price = sqldf("""SELECT partlor_name, district, unique_menu, price 
                        FROM burger 
                        WHERE price BETWEEN 50000 AND 100000""")
burger_price.style.hide_index()
```

```python
#below IDR 100000
burger_price = sqldf("""SELECT partlor_name, district, unique_menu, price 
                        FROM burger 
                        WHERE price <= 50000""")
burger_price.style.hide_index()
```

# Rating

Here is the best restaurant available in West Jakarta, the highest is Carl's Jr. Central Park and Lippo Mall Puri, each scored 4.3. Followed by Flip Burger Kebon Jeruk which scored 4.1. If the score is more than 3.5 then the partlor is worthed to try. If the score is more than 3.5 then the partlor is worthed to try. All the facility and menu are representated by all both google and platform rating and count, so in the meantime only ratings are counted here.

```python
burger = sqldf("""SELECT partlor_name, district, 
                         rtrim(google_rating) as google_rating, 
                         rtrim(zomato_rating) as zomato_rating, 
                         rtrim(qraved_rating) as qraved_rating, 
                         rtrim((google_rating + zomato_rating + qraved_rating)/3, 1) as total_rating,
                    CASE
                        WHEN ltrim((google_rating + zomato_rating + qraved_rating)/3,1) >= 3.5 THEN "Recommended"
                        ELSE "Reconsider"
                    END AS "Recommendation"
                    FROM burger
                    ORDER BY total_rating DESC""")
burger.style.hide_index()
```
