# Steakhouse Draft Menu List : Western

Western is a common genre in steak area as its 'originated', where the price range are above IDR 200k. They have the most complete course from appetizer, main course, and dessert. 

The poultry also the most vary especially for the beef choices. Which are striploin, chateaubriand, mignon, t-bone, and tomahawk (other than ordinary sirloin-tenderloin-rib eye). Even they have lamb and foie grass as well. For seafood section there are one or two places served barramundi-- all are in South Jakarta.

There are 46 western restaurants in Jakarta, special for this section I divide based on city, but unfortunately East Jakarta can't be displayed as they don't have any western like those restaurant, mostly available on other sections especially Local and Local Westerned. Each section contained of glimpse menu and price, poultry, and courses. There are also facility at the bottom, providing options like takeaway, delivery, outdoor/smoking area, alcohol, wifi, reservation, and other special features.     

```python
import pandas as pd
from pandasql import sqldf
steak = pd.read_csv('../input/steakhouse-jakarta/Steakhouse.csv', sep = ";")
```

```python
steak = sqldf("""SELECT COUNT(specialized) AS 'Western Total' 
                 FROM steak 
                 WHERE specialized = 'Western'""")
steak.style.hide_index()
```

# Prices
## Central Jakarta

In this area, the steakhouse mostly located in Thamrin and Sudirman (office building and shopping mall), least in Menteng and Cikini (owned place). Most choices for beef are rib eye, sirloin, and tenderloin. **SKYE Thamrin** has the most beef available including tomahawk and t-bone. **MUCCA Steak (Sudirman)** has the most expensive menu which is Tomahawk Bone in Rib Eye Steak, priced for IDR 9.400k. Rest of them are ranged from IDR 190k until IDR 480k, averagely IDR 280k.

```python
steak = sqldf("""SELECT restaurant_name, district, place, unique_menu, price
                 FROM steak
                 WHERE city = 'Central Jakarta' AND specialized = 'Western' 
                 ORDER BY price DESC""")
steak.style.hide_index()
```

# Facility
