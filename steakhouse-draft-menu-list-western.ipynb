# Steakhouse Draft Menu List : Western

Western is a common genre in steak area as its 'originated', where the price range are above IDR 200k. They have the most complete course from appetizer, main course, and dessert. 

The poultry also the most vary especially for the beef choices. Which are striploin, chateaubriand, mignon, t-bone, and tomahawk (other than ordinary sirloin-tenderloin-rib eye). Even they have lamb and foie grass as well. For seafood section there are one or two places served barramundi-- all are in South Jakarta.

There are 44 western restaurants in Jakarta, special for this section I divide based on city, but unfortunately East Jakarta can't be displayed as they don't have any western like those restaurant, mostly available on other sections especially Local and Local Westerned. Each section contained of glimpse menu and price, poultry, and courses. There are also facility at the bottom, providing options like takeaway, delivery, outdoor/smoking area, alcohol, wifi, reservation, and other special features.     

```python
import pandas as pd
from pandasql import sqldf
steak = pd.read_csv('../input/steakhouse-jakarta/Steakhouse.csv', sep = ";")
```

```python
steak = sqldf("""SELECT COUNT(specialized) 
FROM steak 
WHERE specialized = 'Western'""")
steak.style.hide_index()
```

# Central Jakarta

In this area, the steakhouse mostly located in Thamrin and Sudirman (office building and shopping mall), least in Menteng and Cikini (owned place). Most choices for beef are rib eye, sirloin, and tenderloin. **SKYE Thamrin** has the most beef available including tomahawk and t-bone. **MUCCA Steak** has the most expensive menu which is Tomahawk Bone in Rib Eye Steak, priced for IDR 9.400k.

As for chicken they are mostly either grilled or roasted. They also have mostly salmon on their seafood menu. Despite this, **Tony Roma's (Thamrin)** and **Gandy Steak House (Menteng)** both have chicken menu like steak and schnitzel, also much more seafood, some platters, and other dish. For other meat they have pork on Tony Roma's and SKYE, lamb on Gandy Steak House, and foie gras for OSCAR (Thamrin) and SKYE.

For appetizer they mostly have salad, pasta, and soup. Also wraps and quesadilla. They also have burger, fries, and pizza. They mostly have ice cream and cake for calling the session. Watch for Tony Roma's and SKYE which has special dessert menu, **OSCAR** for the most sweet menu.

```python
steak = sqldf("""SELECT restaurant_name, district, place, unique_menu, price
FROM steak
WHERE city = 'Central Jakarta' AND specialized = 'Western' 
ORDER BY price DESC""")
steak.style.hide_index()
```

```python
steak = sqldf("""SELECT restaurant_name, district, place, beef, chicken, seafood, other_poultry
FROM steak
WHERE city = 'Central Jakarta' AND specialized = 'Western'""")
steak.style.hide_index()
```

```python
steak = sqldf("""SELECT restaurant_name, district, place, appetizer, other_dish, dessert
FROM steak
WHERE city = 'Central Jakarta' AND specialized = 'Western'""")
steak.style.hide_index()
```

# Facility

They have many takeaway options but delivery are rare, very rare. Even they mostly don't have outdoor options but luckily smoking area are much available. Wine, cocktail, beer and liquor are most available alcohol options, almost all of them has wifi and can be reserved. 

There are Gandy Steak House which also turned up to bakery. Mucca Steak and BarBarRibs are claimed to be kids friendly. **J Steak (Blok M)** and **Holy Smokes (Senopati)** have their own valet parking, **SKYE** has their best city view. We can also costumize our order freely at **Super Steak (Blok M)**, meanwhile B'Steak and Street Steak both have payer room, private room are much more accessible here.

```python
steak = sqldf("""SELECT restaurant_name, city, district, takeaway, delivery, outdoor, smoking_area,
alcohol_served, wifi, reservation, other
FROM steak 
WHERE specialized = 'Western'""")
steak.style.hide_index()
```
