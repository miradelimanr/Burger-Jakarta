```python
import pandas as pd
from pandasql import sqldf
burger = pd.read_csv('../input/south-jakarta-burger/south_jakarta.csv', encoding = 'unicode_escape', sep = ";")
```

# District

```python
district = sqldf("SELECT DISTINCT district FROM burger").transpose()
district
```

Then here is the restaurant/partlors available in South Jakarta. There are total 24 vendors and rather dominated by fast food and budget section. Mostly Burger King, Blenger Burger, and Flip Burger. Least fast food available is Carl's Jr and the high end available are Lawless Burgerbar and TGI Fridays Lippo Kemang.

```python
burger = sqldf("SELECT no, partlor_name, district, unique_menu, price FROM burger")
burger.style.hide_index()
```

# Price-Range

We sorted the list from the most expensive and also the cheapest, also with the price range from IDR 50000 to 100000. Here is the order from the most expensive. The most expensive here is TGI Fridays Lippo Mall Kemang which ranged 100k per menu, the cheapest is Burger Bakar Qebul which ranged 13k per menu. Mostly budgeted below 50k.

```python
#most expensive
burger_price = sqldf("SELECT partlor_name, district, unique_menu, price\ 
FROM burger ORDER BY price DESC")
burger_price.head().style.hide_index()
```

```python
#cheapest
burger_price = sqldf("SELECT partlor_name, district, unique_menu, price FROM burger ORDER BY price")
burger_price.head().style.hide_index()
```

```python
#more than IDR 100000
burger_price = sqldf("SELECT partlor_name, district, unique_menu, price\
FROM burger WHERE price >= 100000")
burger_price.style.hide_index()
```

```python
#between 50000 to 100000
burger_price = sqldf("SELECT partlor_name, district, unique_menu, price\
FROM burger WHERE price BETWEEN 50000 AND 100000")
burger_price.style.hide_index()
```

```python
#below IDR 100000
burger_price = sqldf("SELECT partlor_name, district, unique_menu, price\
FROM burger WHERE price <= 50000")
burger_price.style.hide_index()
```

I'll calculate and decide which are recommended on the other file but here is the glimpse. We can see which vendor has the highest rating available. For this time I order this list by google_rating as it appeared first.
