```python
import pandas as pd
from pandasql import sqldf
burger = pd.read_csv('../input/east-jakarta-burger/east_jakarta.csv', sep = ";")
```

# District

```python
partlor_name = sqldf("SELECT DISTINCT district FROM burger").transpose()
partlor_name
```

Here is the list of the restaurant. Consists of 9 restaurant/partlors available.

```python
burger = sqldf("SELECT no, partlor_name, District, unique_menu, price FROM burger")
burger.style.hide_index()
```

# Price-Range

We sorted the list from the most expensive and also the cheapest, also with the price range from IDR 50000 to 100000. Here is the order from the most expensive. The most expensive here is Carl's Jr. Klender which their premium menu is priced about 80k, the cheapest menu is Dons Burger Pusat Grosir Cililitan for at least 13k per menu. 

No prices more than 100k, mostly below 50k per menu.

```python
#most expensive
burger_price = sqldf("SELECT partlor_name, district, unique_menu, price FROM burger ORDER BY price DESC")
burger_price.head().style.hide_index()
```

```python
#cheapest
burger_price = sqldf("SELECT partlor_name, district, unique_menu, price FROM burger ORDER BY price")
burger_price.head().style.hide_index()
```

```python
#more than IDR 100000
burger_price = sqldf("SELECT partlor_name, district, unique_menu, price FROM burger WHERE price >= 100000")
burger_price.style.hide_index()
```

```python
#between 50000 to 100000
burger_price = sqldf("SELECT partlor_name, district, unique_menu, price FROM burger WHERE price BETWEEN 50000 AND 100000")
burger_price.style.hide_index()
```

```python
#below IDR 50000
burger_price = sqldf("SELECT partlor_name, district, unique_menu, price FROM burger WHERE price <= 50000")
burger_price.style.hide_index()
```

Let's take a look for all the facility, I counted each of them also.

```python
burger = sqldf("SELECT no, partlor_name, district, takeaway, delivery, outdoor_seat,\
smoking_area, alcohol_served, wifi, reservation, drive_thru, halal_food FROM burger")
burger.head().style.hide_index()
```

# Rating
I'll calculate and decide which are recommended on the other file but here is the glimpse. We can see which vendor has the highest rating available. For this time I order this list by google_rating as it appeared first.

```python
rating = sqldf("SELECT partlor_name, district, ltrim(google_rating) as google_rating,\
ltrim(zomato_rating) as zomato_rating, ltrim(qraved_rating) as qraved_rating,\
zomato_count, qraved_count FROM burger ORDER BY google_rating DESC")
rating.style.hide_index()
```
